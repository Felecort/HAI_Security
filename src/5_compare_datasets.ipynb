```python
import pandas as pd
from catboost import Pool, CatBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, f1_score, accuracy_score
```

```python
dataset_1 = pd.read_pickle('../data/prepared_data/train_1_fi.pkl')
dataset_2 = pd.read_pickle('../data/prepared_data/train_2_fi.pkl')
```

```python
dataset_1['dataset_label'] = 0
dataset_2['dataset_label'] = 1
```

```python
dataset = pd.concat([dataset_1, dataset_2])
```

```python
dataset.drop(columns=['Timestamp', 'label'], inplace=True)
```

```python
train_set, test_set = train_test_split(dataset, test_size=0.5, shuffle=True, random_state=42)
train_pool = Pool(train_set.drop(columns='dataset_label'),
                  train_set['dataset_label'])
test_pool = Pool(test_set.drop(columns='dataset_label'))
```

```python
model = CatBoostClassifier(thread_count=-1)
model.fit(train_pool)
```

```python
preds = model.predict(test_pool)
```

```python
print(roc_auc_score(test_set['dataset_label'], preds))
print(accuracy_score(test_set['dataset_label'], preds))
print(f1_score(test_set['dataset_label'], preds))
```
